```python
import pandas as pd
import random

```

```python
years = list(range(2010, 2025))
states = [
    "Maharashtra", "Delhi", "Uttar Pradesh", "Karnataka",
    "Tamil Nadu", "Gujarat", "Rajasthan",
    "Madhya Pradesh", "West Bengal", "Bihar"
]

crime_types = [
    "Murder", "Rape", "Theft", "Robbery",
    "Assault", "Cyber Crime",
    "Kidnapping", "Domestic Violence"
]

data = []

for year in years:
    for state in states:
        for crime in crime_types:
            crime_count = random.randint(50, 5000)
            data.append([year, state, crime, crime_count])

df = pd.DataFrame(data, columns=[
    "Year", "State", "Crime_Type", "Crime_Count"
])

df.to_csv("crime_data.csv", index=False)

print("✅ crime_data.csv generated with", len(df), "rows")
```
