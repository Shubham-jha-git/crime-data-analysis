```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("../data/processed/crime_data_cleaned.csv")
df.head()
```

```python
yearly_crime = df.groupby("Year")["Crime_Count"].sum()

plt.figure()
yearly_crime.plot()
plt.title("Total Crimes per Year (India)")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.show()
```

```python
state_crime = df.groupby("State")["Crime_Count"].sum().sort_values(ascending=False)
state_crime
```

```python
plt.figure()
state_crime.plot(kind="bar")
plt.title("Total Crimes by State")
plt.xlabel("State")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.show()
```

```python
crime_type_count = df.groupby("Crime_Type")["Crime_Count"].sum().sort_values(ascending=False)
crime_type_count
```

```python
plt.figure()
crime_type_count.plot(kind="bar")
plt.title("Total Crimes by Crime Type")
plt.xlabel("Crime Type")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.show()
```
